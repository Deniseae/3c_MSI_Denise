```python
import pandas as pd
import os
from google.colab import drive

drive.mount('/content/drive')
os.listdir("/content/drive/MyDrive/Dataset")
df = pd.read_csv('/content/drive/MyDrive/Dataset/VENTAS.csv')
df.head()
```

```python
#lista de diccionarios con productos
ventas_lista = [
    {"producto": "Heladera", "precio": 680000, "cantidad": 4},
    {"producto": "Lavarropa", "precio": 750000, "cantidad": 1},
    {"producto": "Televisor", "precio": 400000, "cantidad": 3}
]

# Calcular total de cada producto
for v in ventas_lista:
    total = v["precio"] * v["cantidad"]
    print(f"{v['producto']}: ${total}")
```

```python
# Primeras filas
df.head()

# Información general
df.info()

# Estadísticas básicas
df.describe()

# Columnas del DataFrame
print("Columnas:", df.columns.tolist())

# Valores únicos por columna
df.nunique()
```

```python
# Valores nulos
print("Valores nulos por columna:")
print(df.isnull().sum())

# Filas duplicadas
duplicados = df.duplicated().sum()
print(f"Filas duplicadas: {duplicados}")

# Mostrar duplicados si existen
if duplicados > 0:
    display(df[df.duplicated()])

# Documentación del estado inicial
print("\n--- Estado inicial de los datos ---")
print(f"Filas totales: {len(df)}")
print(f"Columnas totales: {len(df.columns)}")
```
