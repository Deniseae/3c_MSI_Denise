```python
#librer√≠as
import pandas as pd
#cargar archivo
from google.colab import files
uploaded = files.upload()
#CSV (usa punto y coma como separador)
df = pd.read_csv('VENTAS.csv', sep=';')
#Mostrar primeras filas
df.head()
```

```python
# Convertir FECHA a tipo datetime
df['FECHA'] = pd.to_datetime(df['FECHA'], dayfirst=True)

# Crear columna con el mes
df['MES'] = df['FECHA'].dt.to_period('M')

# Calcular total de cada venta
df['TOTAL_VENTA'] = df['CANTIDAD'] * df['PRECIO']

# Agrupar por mes
ventas_mensuales = df.groupby('MES')['TOTAL_VENTA'].sum().reset_index()

# Mostrar resultados
print("Ventas mensuales:")
print(ventas_mensuales)
```

```python
#lista de diccionarios con productos
ventas_lista = [
    {"producto": "Heladera", "precio": 680000, "cantidad": 4},
    {"producto": "Lavarropa", "precio": 750000, "cantidad": 1},
    {"producto": "Televisor", "precio": 400000, "cantidad": 3}
]

# Calcular total de cada producto
for v in ventas_lista:
    total = v["precio"] * v["cantidad"]
    print(f"{v['producto']}: ${total}")
```
